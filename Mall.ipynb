# Mall Exploratory Data Analysis

###### This data is about weekly sales at 26 Walmart stores from 2010 to 2012.

## Load Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("Mall.xlsx")
```

## Data Extraction

```python
print(df.info())
```

```python
print(df.describe())
```

## Data Cleaning

Data is cleaned!

## Calculation Datas

```python
for col in df.columns:
    print("minimum", col, "is :", df[col].min())

print("\n")

for col in df.columns:
    print("maximum", col, "is :", df[col].max())

print("\n")

for col in df.columns:
    print("average", col, "is :", df[col].mean())

print("\n")

for col in df.columns:
    print("standard deviation", col, "is :", df[col].std())
```

## Compair All Stores In Two Plots

```python
total_sales_list = []

for store_id in range(1, 14):
    store_sales = df[df["Store"] == store_id]["Sales"]
    total_sales = store_sales.sum()
    total_sales_list.append(total_sales)

fig, ax = plt.subplots()
stores = [str(i) for i in range(1, 14)]
sales = total_sales_list
bar_labels = [""] * len(stores)
bar_colors = ["skyblue"] * len(stores)
ax.bar(stores, sales, label=bar_labels, color=bar_colors)
ax.set_ylabel("Total Sales")
ax.set_title("Total Sales by Store")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
total_sales_list = []

for store_id in range(14, 27):
    store_sales = df[df["Store"] == store_id]["Sales"]
    total_sales = store_sales.sum()
    total_sales_list.append(total_sales)

fig, ax = plt.subplots()
stores = [str(i) for i in range(14, 27)]
sales = total_sales_list
bar_labels = [""] * len(stores)
bar_colors = ["red"] * len(stores)
ax.bar(stores, sales, label=bar_labels, color=bar_colors)
ax.set_ylabel("Total Sales")
ax.set_title("Total Sales by Store")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

 نتیجه گیری : این دو نمودار به طور کلی فروش هفتگی هر فروشگاه را نشان می دهد و طبق این نمودار فروشگاه های 4.13.14.20 از بیشترین فروش هفتگی برخوردارند و همینطور طبق نمودار فروشگاه های 3.5.26 از کمترین فروش هفتگی برخوردارند (شاید این امار می تواند از عناصری مانند رفتار پرسنل . مکان احداث فروشگاه . محصولات موجود در فروشگاه و ... بستگی داشته باشد)

Description : These two charts generally show the weekly sales of each store, and according to this chart, stores 4.13.14.20 have the highest weekly sales, and stores 3.5.26 have the lowest weekly sales, according to the chart (perhaps this statistic could depend on elements such as personnel behavior, store location, products available in the store, etc.)

## Compair All Stores In One Plot

```python
half_stores1 = df[(df["Store"] >= 1) & (df["Store"] <= 13)]
half_stores2 = df[(df["Store"] >= 14) & (df["Store"] <= 26)]

total_sales1 = half_stores1["Sales"].sum()
total_sales2 = half_stores2["Sales"].sum()


fig, ax = plt.subplots()

stores = ["1-13","14-26"]
sales = [total_sales1, total_sales2]
bar_labels = ["red", "blue"]
bar_colors = ["tab:red", "tab:blue"]

ax.bar(stores, sales, label=bar_labels, color=bar_colors)

ax.set_ylabel("Weekly Sales")
ax.set_title("Total Sales by Store")
ax.legend()

plt.show()
```

توضیحات : این نمودار مقایسه 13 فروشگاه اول با 13 فروشگاه دوم از لحاظ فروش هفتگی است و با توجه به این نمودار میتوان نتیجه گرفت که فروش هفتگی در 13 فروشگاه اول بیشتر از 13 فروشگاه دوم است

Description : This chart compares the first 13 stores with the second 13 stores in terms of weekly sales. According to this chart, it can be concluded that weekly sales in the first 13 stores are higher than in the second 13 stores.

## Compair Holiday And Working Day

```python
holiday_date = df[df["Holiday"].isin([True])]
working_day = df[df["Holiday"].isin([False])]

holiday_sales = holiday_date["Sales"].sum()
working_day_sales = working_day["Sales"].sum()

labels = "HOLIDAY", "WORKING DAY"
sizes = [holiday_sales, working_day_sales]

fig, ax = plt.subplots()

explode = (0.1, 0.1)
ax.pie(sizes, explode= explode, labels = labels, autopct ="'%1.1f%%", startangle = 90)
plt.show()
```

توضیحات : این نمودار مقایسه فروش هفتگی روز های تعطیل در برابر روز های غیر تعطیل را نمایش می دهد و با توجه به این نمودار می توانیم نتیجه بگیریم که روز های غیر تعطیل نقش بسیار زیادی را در فروش دارند

Description : This chart shows the comparison of weekly sales on holidays versus non-holidays, and based on this chart, we can conclude that non-holidays play a very large role in sales.

## Compair The Average Holiday And Working Day

```python
mean_sales_holiday = df[df["Holiday"] == True]["Sales"].mean()
mean_sales_working_day = df[df["Holiday"] == False]["Sales"].mean()


print(f'Average Sales on Holidays: {mean_sales_holiday}')
print(f'Average Sales on Working Day: {mean_sales_working_day}')


labels = ["Holidays", "Working Day"]
sales_means = [mean_sales_holiday, mean_sales_working_day]

plt.figure(figsize=(10, 6))
plt.bar(labels, sales_means, color=["blue", "green"])

plt.title("Comparison of Average Sales on Holidays and Working Day")
plt.show()
```

توضیحات : این نمودار هم مقایسه میانگین فروش هفتگی در روزهای تعطیل و غیر تعطیل است و می توان نتیجه گرفت که علارغم اینکه در نمودار قبلی نقش قابل توجهی روزهای غیر تعطیل داشت ولی در این نمودار خلاف نمودار قبلی میانگین فروش هفتگی روزهای تعطیل بیشتر از روزهای غیر تعطیل بود

Description: This chart compares the average weekly sales on holidays and non-holidays, and it can be concluded that despite the fact that non-holidays played a significant role in the previous chart, in this chart, unlike the previous chart, the average weekly sales on holidays were higher than on non-holidays.

## Compair Four Top Stores

```python
total_sales_list = []

for store_id in 4,13,14,20:
    store_sales = df[df["Store"] == store_id]["Sales"]
    total_sales = store_sales.sum()
    total_sales_list.append(total_sales)


fig, ax = plt.subplots()
season = ["Store 4", "Store 13", "Store 14", "Store 20"]
colors = ["silver", "lightblue", "darkorange", "gold"]
y_pos = np.arange(len(season))
error = np.random.rand(len(season))

ax.barh(y_pos, total_sales_list, xerr = error, align = "center", color = colors)
ax.set_yticks(y_pos, labels=season)
plt.grid(axis="x", linestyle="--", alpha=0.7)
ax.invert_yaxis() 
plt.show()
```

توضیحات : بعد از اینکه نمودار های تمامی فروشگاه ها را رسم کردیم نوبت به این رسید تا 4 فروشگاه اول را از لحاظ فروش هفتگی با هم مقایسه کنیم تا بهترین فروشگاه را از لحاظ فروش هفتگی پیدا کنیم که با توجه به این نمودار می توانیم نتیجه بگیریم که فروشگاه شماره 20 با اختلاف بسیار کمی از فروشگاه شماره 4 رتبه اول فروشگاه ها را از لحاظ فروش هفتگی توانست کسب کند

Description : After we drew the graphs of all the stores, it was time to compare the first 4 stores in terms of weekly sales to find the best store in terms of weekly sales. According to this graph, we can conclude that store number 20 was able to achieve the first rank of stores in terms of weekly sales by a very small margin from store number 4.

## Compair Three Down Stores

```python
total_sales_list = []

for store_id in 3,5,26:
    store_sales = df[df["Store"] == store_id]["Sales"]
    total_sales = store_sales.sum()
    total_sales_list.append(total_sales)


fig, ax = plt.subplots()
season = ["Store 3", "Store 5", "Store 26"]
colors = ["darkorange", "silver", "gold"]
y_pos = np.arange(len(season))
error = np.random.rand(len(season))

ax.barh(y_pos, total_sales_list, xerr = error, align = "center", color = colors)
ax.set_yticks(y_pos, labels=season)
plt.grid(axis="x", linestyle="--", alpha=0.7)
ax.invert_yaxis() 
plt.show()
```

توضیحات : همان طور که 4 فروشگاه برتر را مشخص کردیم حال برای تعیین 3 فروشگاه با عملکرد ضعیف هم این کار را انجام دادیم که با توجه به داده هایمان به این نتیجه رسیدیم که فروشگاه شماره 26 بدترین عملکرد از لحاظ فروش هفتگی را دارد

Description : Just as we identified the top 4 stores, we also did the same to identify the 3 poorly performing stores. Based on our data, we concluded that store number 26 had the worst performance in terms of weekly sales.

## Weekly Sales For First Month Of Each Store

```python
date = np.array(df["Date"].iloc[0:4])


store1 = df[df["Store"] == 1]
sum_store1 = store1.head(4)["Sales"]

store2 = df[df["Store"] == 2]
sum_store2 = store2.head(4)["Sales"]

store3 = df[df["Store"] == 3]
sum_store3 = store3.head(4)["Sales"]

store4 = df[df["Store"] == 4]
sum_store4 = store4.head(4)["Sales"]

store5 = df[df["Store"] == 5]
sum_store5 = store5.head(4)["Sales"]

plt.figure(figsize=(18, 6))
plt.plot(date, sum_store1, color = "orange", label = "First Store")
plt.plot(date, sum_store2, color = "lightblue", label = "Second Store")
plt.plot(date, sum_store3, color = "red", label = "Third Store")
plt.plot(date, sum_store4, color = "blue", label = "4th Store")
plt.plot(date, sum_store5, color = "green", label = "5th Store")


plt.title("Weekly sales for the first four weeks of stores")
plt.xlabel = ("Date")
plt.ylabel = ("Weekly Sales")
plt.legend()

plt.grid()

plt.show()
```

توضیحات : این نمودار فروش هفتگی 5 فروشگاه اول در اولین ماه را نشان می دهد و با توجه به این نمودار می توان نتیجه گرفت که هفته دوم به دلیل وجود تعطیلی فروشگاه ها فروش هفتگی خوبی را کسب کردند ولی در سه هفته باقی مانده از ماه چون هفته تعطیل نبود به همین خاطر فروش کمتری نسبت به هفته تعطیل داشتند

Description : This chart shows the weekly sales of the first 5 stores in the first month. According to this chart, we can conclude that the second week had good weekly sales due to the stores being closed, but in the remaining three weeks of the month, since there was no closed week, they had lower sales than the closed week.

## Correlation For Datas

```python
correlation_matrix = df[["Store", "Part", "Sales", "Holiday"]].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot = True,cmap = "YlGnBu", center = 0)
plt.title("Correlation between all datas")
plt.show()
```

 توضیحات : تمامی ضرایب همبستگی بین متغیرها بسیار نزدیک به صفر هستند و هیچ همبستگی قوی ای بین متغیرها مشاهده نشد پس طبق مشاهدات نتیجه میگیرم که این متغیرها به صورت مستقل از یکدیگر عمل می‌کنند و تأثیر زیادی بر یکدیگر ندارند 

Description : All correlation coefficients between variables are very close to zero and no strong correlations were observed between variables, so based on the observations, I conclude that these variables act independently of each other and do not have much effect on each other.
